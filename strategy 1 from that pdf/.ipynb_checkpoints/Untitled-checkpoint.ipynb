```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

stonks = []
stonkNames = []

import glob
for file in glob.glob("*.csv"):
    stonkNames.append(file[0:-4])
    stonks.append(pd.read_csv(file)["Open"])

testStonk = stonks[0]

len(testStonk)
```

```python
T = 12   # number of months
S = len(stonks[0]) - (T+1)    # month to start at, inclusive

stonkRankings = [];

for stonk in stonks:
    R = (stonk / stonk.shift(1)) - 1
    print(R.size - (S+T))
    print(R.size - S)
    R = R[R.size - (S+T):R.size - S]
    R = R.reset_index(drop=True)
    rMean = R.sum() / T
    sigma = R.std()
    rRiskAdj = rMean / sigma

    stonkRankings.append(rRiskAdj)
```

```python
for i in range(len(stonks)):
    print(stonkNames[i] + " " + str(stonkRankings[i]))

print(stonkNames[stonkRankings.index(max(stonkRankings))])
```

```python
for i in range(len(stonks)):
    print(stonkNames[i] + " " + str(stonks[i].iloc[T + 1] / stonks[i][T]))
```

```python
for i in range(len(stonks)):
    plt.plot(stonks[i][T:S+T].reset_index(drop=True) / stonks[i][T])
plt.legend(stonkNames)
```

```python
T = 12   # number of months
S = len(stonks[0]) - (T+1)    # month to start at, inclusive

stonkRankings = [];
bestStonks = [];
idk = pd.DataFrame()

for i in range(len(stonks)):
    for j in range(S):
        R = (stonks[i] / stonks[i].shift(1)) - 1
        #print(R.size - (S+T) + j)
        #print(R.size - S + j)
        R = R[R.size - (S+T) + j:R.size - S + j]
        R = R.reset_index(drop=True)
        rMean = R.sum() / T
        sigma = R.std()
        rRiskAdj = rMean / sigma

        stonkRankings.append(rRiskAdj)
    #bestStonks.append(stonkNames[stonkRankings.index(max(stonkRankings))])
    
    idk[stonkNames[i]] = stonkRankings
    
    plt.plot(stonkRankings)
    stonkRankings = [];
plt.legend(stonkNames)
#bestStonks
#print(idk)
idk["predicted"] = idk.idxmax(axis=1)
idk
```

```python
blah = pd.DataFrame()
for i in range(len(stonks)):
    blah[f"{stonkNames[i]}%"] = stonks[i][T:S+T].reset_index(drop=True) / stonks[i][T:S+T].reset_index(drop=True).shift(1)
    plt.plot(blah[f"{stonkNames[i]}%"])
    
idk["actual"] = blah.idxmax(axis=1).str.replace("%", "")
plt.legend(stonkNames)

arr = np.argsort(-idk.iloc[:,:len(stonks)].values, axis=1)
hopethisworksPredicted = pd.DataFrame(idk.columns[arr], index=idk.index)
print(hopethisworksPredicted)

arr = np.argsort(-idk.iloc[:,len(stonks)+1:2*len(stonks)+1].values, axis=1)
hopethisworksActual = pd.DataFrame(idk.columns[arr], index=idk.index)
print(hopethisworksActual)
```

```python
print("Prediction was correct " + str(100 * (idk["predicted"] == idk["actual"]).sum() / (idk.shape[0] - 1)) + "% of the time")
print("Prediction was almost correct " + str(100 * (hopethisworksActual[1] == idk["predicted"]).sum() / (idk.shape[0] - 1)) + "% of the time")
print("Prediction must be correct over " + str(100 / len(stonks)) + "% of the time for first place to be meaningful or " + str(200 / len(stonks)) + "% of the time for two to be meaningful")
```
